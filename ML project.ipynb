```python
##Data preprocessing
```

```python
Churn_Modelling.head(20)
```

```python
gendermapping={'Male':1, 'Female':0}

Churn_Modelling['Gender']=Churn_Modelling['Gender'].map(gendermapping)
```

```python
Churn_Modelling
```

```python
Geomapping={'France':'F', 'Spain': 'S', 'Germany':'G'}

Churn_Modelling['Geography']=Churn_Modelling['Geography'].map(Geomapping)
```

```python
Churn_Modelling
```

```python
pd.get_dummies(Churn_Modelling['Gender'], drop_first=True)
```

```python
sex=pd.get_dummies(Churn_Modelling['Gender'], drop_first=True)
```

```python
sex
```

```python
sex.head()
```

```python
Churn_Modelling= pd.concat([Churn_Modelling,sex], axis=1)
```

```python
Churn_Modelling
```

```python
Churn_Modelling.drop(1, axis=1, implace=True)
```

```python
Churn_Modelling
```

```python
sns.countplot(x='Exited', hue="Gender",data=Churn_Modelling)
```

```python
from sklearn.preprocessing import LabelEncoder
lableEn=LabelEncoder()
Churn_Modelling_tf=Churn_Modelling.copy()

for i in Churn_Modelling.columns:
    Churn_Modelling_tf[i]=lableEn.fit_transform(Churn_Modelling[i])

```

```python
Churn_Modelling_tf
```

```python
X=Churn_Modelling_tf.drop('Exited', axis=1)
Y=Churn_Modelling_tf['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X, Y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape, '---train f shape')
print(X_test.shape, '---test f shape')
print(Y_train.shape, '---train  Lshape')
print(Y_test.shape, '---test L shape')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Churn_Modelling_model = LogisticRegression()
```

```python
Churn_Modelling_model.fit(X_train, Y_train)
```

```python
prediction = Churn_Modelling_model.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test, prediction))
```

```python
##acurcy score
```

```python
import sklearn.metrics as sm
```

```python
acc_model=sm.accuracy_score(Y_test, prediction)
print('Accuracy score ===', str=(acc_model))
```
